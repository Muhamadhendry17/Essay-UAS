```python
def higherThan80(data):
    result = []
    
    for item in data:
        if item.get("grade") > 80:
            result.append(item)
            
    return result
```

```python
def printData(data):
    for item in data:
        print("=====================")
        print("Student Number: ", item.get("student number"))
        print("Name: ", item.get("name"))
        print("Subject: ", item.get("subject"))
        print("Grade: ", item.get("grade"))
        print("=====================")
        print("\n")
```

```python
print("Data dengan nilai lebih dari 80\n")
printData(higherThan80(data))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.DataFrame(data)
dataFrame = dataFrame.set_index("student number")
dataFrame
```

```python
dataFrame["description"] = ["FAIL" if grade < 70 else "PASS" for grade in dataFrame["grade"]]
dataFrame
```

```python
groupedDataFrame = dataFrame.groupby(["student number", "name"])["grade"].mean()
groupedDataFrame.set_index("name")
```

```python
groupedDataFrame.plot(kind="bar")
plt.xlabel("Name")
plt.ylabel("Grade")
plt.title("Mean of Student Grades Values")
plt.show()
```
